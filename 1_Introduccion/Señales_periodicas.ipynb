## Frecuencia

```python
import numpy as np
import matplotlib.pyplot as plt

# Vector tiempo
tiempo = np.linspace(0, 1, 100)      # Genera un array de 100 elementos entre 0 y 1
                                     # Es decir, valores equiespaciados entre el intervalo deseado
                                     # Con el range de python, se generan enteros solamente. Entre otras ventajas
frecuencia = 2

# w = 2 * pi * frecuencia
senial_t = np.sin(2 * np.pi * frecuencia * tiempo)
                                # C√≥mo variable dependiente, el seno de los valores
                                # Se aplica elemento a elemento. M√©todo de numpy y sus ventajas.
```

```python
# Crear el gr√°fico
plt.figure(figsize=(8, 6))    # Para definir un tama√±o personalizado de la gr√°fica. No es necesario
plt.plot(tiempo, senial_t, label='s(t) = sen(w*t)')
plt.title('Gr√°fica de la se√±al')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

## Amplitud + Frecuencia

```python
def senoidal(amplitud, frecuencia, tiempo=1):
    """
    Genera una se√±al senoidal.
    Esta funci√≥n genera una se√±al senoidal en funci√≥n de los par√°metros especificados.

    Par√°metros:
    ----------
    amplitud : float
        La amplitud de la se√±al senoidal.
    frecuencia : float
        La frecuencia de la se√±al senoidal en hertz (Hz).
    tiempo : float, opcional
        La duraci√≥n total de la se√±al en segundos. Por defecto, es 1 segundo.

    Retorna:
    -------
    tuple
        Una tupla que contiene dos arrays NumPy: el primero es el array de tiempo y el segundo es el array de valores de la se√±al senoidal.

    Ejemplos:
    --------
    # Generar una se√±al senoidal con amplitud 2, frecuencia 5 Hz y duraci√≥n de 2 segundos
    tiempo, senial = senoidal(2, 5, 2)

    # Generar una se√±al senoidal con amplitud 1.5, frecuencia 10 Hz y duraci√≥n de 0.5 segundos
    tiempo, senial = senoidal(1.5, 10, 0.5)
    """
    tiempo = np.linspace(0, tiempo, 400)
    w = 2 * np.pi * frecuencia
    senial_t = amplitud * np.sin(w * tiempo)  # Amplitud x seno
    return tiempo, senial_t
```

```python
t1, s1 = senoidal(amplitud=2, frecuencia=2)

plt.figure(figsize=(8, 6))    # Para definir un tama√±o personalizado de la gr√°fica. No es necesario
plt.plot(t1, s1, label='Se√±al 1')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
t1, s1 = senoidal(amplitud=2, frecuencia=2)
t2, s2 = senoidal(amplitud=1, frecuencia=2)

plt.figure(figsize=(8, 6))    # Para definir un tama√±o personalizado de la gr√°fica. No es necesario
plt.plot(t1, s1, label='Se√±al 1')
plt.plot(t2, s2, label='Se√±al 2')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
t1, s1 = senoidal(amplitud=2, frecuencia=2)
t2, s2 = senoidal(amplitud=1, frecuencia=4)

plt.figure(figsize=(8, 6))    # Para definir un tama√±o personalizado de la gr√°fica. No es necesario
plt.plot(t1, s1, label='Se√±al 1')
plt.plot(t2, s2, label='Se√±al 2')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

## Interactivo

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def plot_periodic_signal(duration, frequency, amplitude):
    t, signal = senoidal(amplitud=amplitude, frecuencia=frequency, tiempo=duration)

    plt.figure(figsize=(10, 6))
    plt.plot(t, signal)
    plt.title(f"Se√±al senoidal peri√≥dica: Frecuencia = {frequency} Hz, Amplitud = {amplitude}")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.grid(True)
    plt.show()
```

```python
duration_slider = widgets.FloatSlider(value=1.0, min=0.0, max=10.0, step=0.5, description='Duraci√≥n (s)')
frequency_slider = widgets.FloatSlider(value=1.0, min=0.0, max=10.0, step=0.5, description='Frecuencia (Hz)')
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.0, max=5.0, step=0.5, description='Amplitud')

interactive_plot = interactive(
    plot_periodic_signal,
    duration=duration_slider,
    frequency=frequency_slider,
    amplitude=amplitude_slider
)

output = interactive_plot.children[-1]
output.layout.height = '600px'
interactive_plot
```

## RMS

```python
t, senial = senoidal(amplitud=1.5, frecuencia=1)

senial_RMS = senial*senial

plt.figure(figsize=(8, 6))
plt.plot(t, senial, label='Se√±al base')
plt.plot(t, senial_RMS, label='Se√±al RMS')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

## Valor Medio

```python
t , senial = senoidal(amplitud=1, frecuencia=2)
senial_cc = senial + 0.2
```

```python
plt.figure(figsize=(8, 6))
plt.plot(t, senial, label='Se√±al base')
plt.plot(t, senial_cc, label='Se√±al CC')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('S(t)')
plt.legend()
plt.grid(True)
plt.show()
```

# Se√±al Cuadrada

```python
def cuadrada(amplitud, frecuencia, tiempo=1, duty_cycle=0.5):
    """
    Genera una se√±al cuadrada.
    Esta funci√≥n genera una se√±al cuadrada en funci√≥n de los par√°metros especificados.

    Par√°metros:
    ----------
    amplitud : float
        La amplitud de la se√±al cuadrada.
    frecuencia : float
        La frecuencia de la se√±al cuadrada en hertz (Hz).
    tiempo : float, opcional
        La duraci√≥n total de la se√±al en segundos. Por defecto, es 1 segundo.
    duty_cycle : float, opcional
        El ciclo de trabajo de la se√±al cuadrada, en el rango de 0.0 a 1.0. Por defecto, es 0.5.

    Retorna:
    -------
    tuple
        Una tupla que contiene dos arrays NumPy: el primero es el array de tiempo y el segundo es el array de valores de la se√±al cuadrada.

    Ejemplos:
    --------
    # Generar una se√±al cuadrada con amplitud 2, frecuencia 5 Hz y duraci√≥n de 2 segundos
    tiempo, senial = cuadrada(2, 5, 2)

    # Generar una se√±al cuadrada con amplitud 1.5, frecuencia 10 Hz, duraci√≥n de 0.5 segundos y ciclo de trabajo 0.3
    tiempo, senial = cuadrada(1.5, 10, 0.5, 0.3)
    """
    tiempo = np.linspace(0, tiempo, 200)  # Crear un array de tiempo
    periodo = 1 / frecuencia  # Calcular el per√≠odo
    num_pulsos = int(tiempo[-1] / periodo)  # Calcular el n√∫mero de pulsos en el tiempo dado
    pulsos = np.zeros_like(tiempo)  # Crear un array para almacenar los valores de los pulsos

    for i in range(num_pulsos):
        t_inicio = i * periodo
        t_fin = t_inicio + periodo * duty_cycle
        pulsos[(tiempo >= t_inicio) & (tiempo <= t_fin)] = amplitud

    return tiempo, pulsos
```

```python
t1, s1 = cuadrada(amplitud=1, frecuencia=5)

plt.figure(figsize=(8, 6))    # Para definir un tama√±o personalizado de la gr√°fica. No es necesario
plt.plot(t1, s1, label='Se√±al 1')
plt.title('Gr√°fica de las se√±ales')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('C(t)')
plt.legend()
plt.grid(True)
plt.show()
```
